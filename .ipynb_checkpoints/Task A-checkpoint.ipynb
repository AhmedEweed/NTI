```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler,Normalizer
```

```python
data=pd.read_csv("dataset.csv")
data.tail()
```

```python
data.info()
```

# Get a Copy of the data to work on and clean

```python
C_data = data.copy()
```

# Feature Engineering


```python
C_data.drop("customerID",axis=1,inplace=True)
```

# Total Charges categories

```python
C_data.describe()
```

# Divide data into bins and then concatenate them to our data

```python
bins = [18, 402, 1400, 3786, 8684.8]
labels = ['1st Quarter', '2nd Quarter', '3rd Quarter', '4th Quarter']
category = pd.cut(C_data.TotalCharges,bins, labels = labels)
category = category.to_frame()
category.columns = ['TotalChargesRange']

C_data = pd.concat([C_data,category],axis = 1)
```

# New Labels plot

```python
sns.countplot(x = 'TotalChargesRange', data = C_data, palette = 'hls')
plt.xticks(rotation = 90)
plt.show()
```

# Discover missing values 

```python
C_data.isnull().sum()
```

# Getting values to fill the null values with

```python
tenure_mean = C_data["tenure"].mean()
senior_mode = C_data["SeniorCitizen"].mode()[0]
print("Tenure mean:",tenure_mean, '\n' + 'Senior Mode:', senior_mode)
```

# Filling null values

```python
C_data["tenure"].fillna(tenure_mean,inplace=True)
C_data["SeniorCitizen"].fillna(senior_mode,inplace=True)
```

# Checking that what we have done is done right

```python
C_data.isnull().sum()
```

# First we have to convert the newly created category column to object

```python
C_data['TotalChargesRange'] = C_data['TotalChargesRange'].astype('object')
```

# Label Encoding

```python
encodings = {}
for col in C_data.columns:
    if (C_data[col].dtype == "object"):
        encodings[col] = LabelEncoder()
        C_data[col]   =encodings[col].fit_transform(C_data[col])
```

# Standardization

```python
sc_tenure = StandardScaler()
sc_total = StandardScaler()
data["tenure"] = sc_tenure.fit_transform(data["tenure"].values.reshape(-1,1))
data["TotalCharges"] = sc_total.fit_transform(data["TotalCharges"].values.reshape(-1,1))
```
